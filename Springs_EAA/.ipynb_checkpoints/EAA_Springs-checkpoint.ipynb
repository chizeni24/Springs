```python
def data_reader_Comal(file_name,base_dir, skip_rows):
    df      = os.path.join(base_dir,file_name)
    df_read = pd.read_csv(df, thousands=r',', skiprows = skip_rows) 
    return df_read
```

```python
#--------------------------------------------------------------------------------------------------------------------
# ----  simple statistics
#--------------------------------------------------------------------------------------------------------------------

def simple_stat(df_data):
    #print('Number of measurements :   ', len(df_data),'\n')     
    minval = df_data.min()
    maxval = df_data.max()
    # concat them into a DataFrame
    df_stat = pd.concat([minval, maxval], axis = 1)
    # change column heading of the dateframe, df_stat
    df_stat = df_stat.rename(columns={0 : 'Min', 1 : 'Max'})
    return df_stat
```

```python
#---------------------------------------------------------------------------------------
# Data Quality Check -- based on simple statistics 
#---------------------------------------------------------------------------------------

stat_ComalSpring = simple_stat(Comal_Spring_Flow)
stat_ComalSpring.style.set_properties(**{'width': '200px'})
display(stat_ComalSpring)
print('')

stat_SanMarcos_Spring = simple_stat(SanMarcos_Spring_Flow)
stat_SanMarcos_Spring.style.set_properties(**{'width': '200px'})
display(stat_SanMarcos_Spring)
print('')

stat_Hueco_Spring = simple_stat(Hueco_Spring_Flow)
stat_Hueco_Spring.style.set_properties(**{'width': '200px'})
display(stat_Hueco_Spring)
print('')

stat_Leona_Spring = simple_stat(Leona_Spring_Flow)
stat_Leona_Spring.style.set_properties(**{'width': '200px'})
display(stat_Leona_Spring)
print('')


stat_NuecesR_UV_Spring = simple_stat(NuecesR_UV_Spring_Flow)
stat_NuecesR_UV_Spring.style.set_properties(**{'width': '200px'})
display(stat_NuecesR_UV_Spring)
print('')
```

```python
# -------------------------------------------------------------------------------------------------------
# report lake and hydroclimate data 
#--------------------------------------------------------------------------------------------------------
def plot_data(ax, date_time, weather_data, ylabel, y_min, y_max, marker, fontsize):
    ax.plot(date_time, weather_data, marker)
    ax.set_ylabel(ylabel, fontsize  = fontsize)
    ax.set_xlabel('Year', fontsize  = fontsize)
    ax.set_ylim(y_min, y_max)
    ax.tick_params(axis = 'both', which = 'major', labelsize =  fontsize - 2)    
    plt.xticks(rotation = 30)
```

```python
fig      = plt.figure()
fig.set_size_inches(18.5, 26.5)
m        =  5
n        =  2
fontsize = 14
marker  = '-b'

ax=fig.add_subplot(m,n,1)
y_label = "Comal Springflow [m$^3$/s]"
ymin    = 0
ymax    = 16
plot_data(ax, Comal_Spring_Flow.Date, Comal_Spring_Flow['Comal_SF(m3)'], y_label, ymin, ymax, marker,fontsize) 
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)   


ax=fig.add_subplot(m,n,2)
y_label = "San Marcos Springflow [m$^3$/s]"
ymin    = 0
ymax    = 16
plot_data(ax,SanMarcos_Spring_Flow.Date, SanMarcos_Spring_Flow['SanMarcos_SF(m3)'], y_label, ymin, ymax, marker,fontsize) 
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)   


ax=fig.add_subplot(m,n,3)
y_label = "Hueco Springflow [m$^3$/s]"
ymin    = 0
ymax    = 4
plot_data(ax,Hueco_Spring_Flow.Date, Hueco_Spring_Flow['Hueco_SF(m3)'], y_label, ymin, ymax, marker,fontsize) 
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)  



ax=fig.add_subplot(m,n,4)
y_label = "Leona Springflow [m$^3$/s]"
ymin    = 0
ymax    = 600
plot_data(ax,Leona_Spring_Flow.Date, Leona_Spring_Flow['Leona_SF(m3)'], y_label, ymin, ymax, marker,fontsize) 
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)  



ax=fig.add_subplot(m,n,5)
y_label = "Nueces UV Springflow [m$^3$/s]"
ymin    = 0
ymax    = 1600
plot_data(ax,NuecesR_UV_Spring_Flow.Date, NuecesR_UV_Spring_Flow['NuecesR_UV_SF(m3)'], y_label, ymin, ymax, marker,fontsize) 
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)  

```
